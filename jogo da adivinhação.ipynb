```python
from random import randint
from time import sleep
computador = randint (0, 10) # comando que faz o computador escolher um número
print('Olá, como vc está?, vamos brincar um pouco?')
print('-=-' *20)
print('Vou pensar em um número entre 0 e 10, tente adivinhar....')
print('-=-' *20)
jogador = int(input('Diga aí, em que número eu pensei? ')) #aqui o usuário irá escolher um número
print('PROCESSANDO......')
sleep(3)
if jogador == computador:
    print('Parabéns vc conseguiu me vencer!')
else:
    print('Eu venci, eu pensei no número {} e não no número {}!'.format(computador, jogador))


```
